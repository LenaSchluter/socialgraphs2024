> * Explain in your own words: What is the the difference between the html page and the wiki-source?

An HTML page is a web document coded in HTML, while a wiki-source is the raw text and markup used within a wiki system to generate web content.


> * What are the various parameters you can set for a query of the wikipedia api?

Key Parameters
* action: Defines the operation (e.g., query, parse, edit).
* format: Specifies the response format (e.g., json, xml).
* titles: The page titles to query (e.g., Albert_Einstein).
* prop: Properties of pages to retrieve (e.g., revisions, links, info).
* list: Retrieves lists of data (e.g., search, categorymembers).
* generator: Generates a list of pages based on criteria (e.g., allpages).
* pageids: Specific page IDs to query.
* rvprop: Specifies revision properties (e.g., content, user).
* redirects: Follows redirects automatically (redirects=1).
* srsearch: Search query for list=search.
* srlimit: Limits the number of search results.
* continue: Handles query continuation for large result sets.

```python
baseurl = "https://en.wikipedia.org/w/api.php?"
action = "action=query"
title = "titles=Trap_music"
content = "prop=revisions&rvprop=content"
dataformat ="format=json"

query = "{}{}&{}&{}&{}".format(baseurl, action, content, title, dataformat)
print(query)
```

```python
wikiresponse = urllib.request.urlopen(query)
wikidata = wikiresponse.read()
wikitext = wikidata.decode('utf-8')

import json
json.loads(wikitext)
```
